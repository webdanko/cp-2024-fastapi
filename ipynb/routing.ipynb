```python
import pandas as pd
import geopandas as gpd
import urllib.request
import json
from shapely.geometry import shape

def get_route_url(start_point, destination_point):
    api_url = "https://router.project-osrm.org/route/v1/walk/"
    geometry = "?geometries=geojson"
    coords_string = "%s,%s;%s,%s" % (start_point[0], start_point[1], destination_point[0], destination_point[1])
    return api_url + coords_string + geometry

gdf =  gpd.GeoSeries();
with open('centroids_q1.geojson', 'r') as f:
    centroids = json.load(f)

destination_point = (37.479482, 55.541021)
for c in centroids['features']:
    start_point = c['geometry']['coordinates']
    route_url = get_route_url(start_point, destination_point)

    with urllib.request.urlopen(route_url) as url:
        data = json.load(url)

    tracks_data = {'type': 'LineString', 'coordinates': data['routes'][0]['geometry']['coordinates']}
    tracks_shape = shape(tracks_data)
    gp = gpd.GeoSeries([tracks_shape])
    gdf = pd.concat([gdf, gp])

gdf
```
